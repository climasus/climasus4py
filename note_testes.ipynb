```python
import climasus as cs

# Equivalente ao pipeline R em uma chamada
rel = cs.sus_pipeline(
    system="SIM-DO",
    uf="AC",
    year=2023,
    lang="pt",
    groups="respiratory",   # filtra grupo de doenças "respiratory"
    age_group=True,         # cria faixas etárias (equiv. create_age_groups = TRUE)
)

# Materializar como DataFrame pandas
df_analise = rel.df()  # ou .to_df() se preferir; .df() funciona com a API do projeto
```

```python
rel = cs.sus_import(system="SIM-DO", uf="AC", year=2023, cache_dir="dados/cache", verbose=True)
```
